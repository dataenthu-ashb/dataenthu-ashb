# *Question 3:*



You have just started a stall at a fair and want to use your newly acquired Python skills to write a program to help with your sales.

i) Write a program that calculates the amount of change you need to return to the customer. Your program should ask the user for the number of products sold and the amount of cash you receive from the customer.
The products you sell have the following prices:

- 1 lemonade = EUR 2.5
- 1 candy = EUR 1.5
- 1 lollipop = EUR 1

To test your program, try it out for a sale of 3 lemonades, 4 candies, and 2 lollipops. \
*Hint: Remember to use the `input()` function to get the user's input and print the resulting change to return.*


ii) Congrats! You have been successfully running your stand 3 days in a row. Display the contribution of each product sold to the total revenue given that you have sold 59 lemonades, 70 candies and 35 lollipops.

```python
# sub-question (i):

# given price of each product
LEMONADE_PRICE = 2.5
CANDY_PRICE = 1.5
LOLLIPOP_PRICE = 1

# get the amount of each product sold
lemonades = input("Lemonades sold: ")
candies = input("Candies sold: ")
lollipops = input("Lollipops sold: ")

# convert the input() results from string to int
lemonades = int(lemonades)
candies = int(candies)
lollipops = int(lollipops)

# calculate the total price of all the products
sum = LEMONADE_PRICE * lemonades + CANDY_PRICE * candies + LOLLIPOP_PRICE * lollipops
# print("Price of all products: " + str(sum))

# The amount of cash received from the customer
money = input("Cash given: ")
money = int(money)

# calculate the difference to return and print
rest = money - sum
print("Amount to return: " +  str(rest))
```

```python
# sub-question (ii):

# given price of each product
# NOTE: you do not need to define this again in a Jupyter Notebook / Colab
LEMONADE_PRICE = 2.5
CANDY_PRICE = 1.5
LOLLIPOP_PRICE = 1

# amount of sales
total_lemonades, total_candies, total_lollipops =  59, 70, 35

# total revenue
total_revenue = total_lemonades * LEMONADE_PRICE + total_candies * CANDY_PRICE + total_lollipops * LOLLIPOP_PRICE

# share of each product to the total revenue
lemonade_share = total_lemonades * LEMONADE_PRICE * 100 / total_revenue
candy_share = total_candies * CANDY_PRICE * 100 / total_revenue
lollipop_share = total_lollipops * LOLLIPOP_PRICE * 100 / total_revenue

print("Lemonade share: " + str(lemonade_share))
print("Candy share: " + str(candy_share))
print("Lollipop share: " + str(lollipop_share))
```
