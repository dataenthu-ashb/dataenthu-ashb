## Question 1: Creating dictionaries

## Question 2: Calculating Demand

Write a function `demand()` to calulate the demanded quantities. The function should take the `price_list` as an argument. Calculate using the formula given below:

$Q_d$ is the total amount of a goods that buyers would choose to purchase at a given price $P$ *as well as* other (exogenous) variables $\bar{Q_d}$, such as income and wealth, prices of substitutes and complements, population, preferences (tastes) and expectations of future prices.
  
$$Q_d = \bar{Q_d} − aP$$

Find the value of $Q_d$ corresponding to the given price in the `price_list` and return a dictionary with the prices from `price_list` as keys and the calculated `Q_d` as values. Also assign the result of your function to dictionary `q_demand`. Your dictionary should be in the form `{2: q_d1, 9: q_d2, 10: q_d3, ...and so on}`

Assume the following values: $a$ = 10, $\bar{Q_d}$ = 1000

```python
# Question 2
a = 10 # sensitivity of the demand to price
Qd_bar = 1000 # exogenous variable, independent qty demanded

# your code here
def demand(price_list):
    q_d = {}
    for price in price_list:
        q_d[price] = Qd_bar - a * price

    return q_d

q_demand = demand(price_list)
q_demand
```

## Question 3: Calculating Supply

Write a function `supply()` to calulate the quantities supplied. Your function should take the `price_list` as an argument. Calculate using the formula given below:

$Q_s$ is the total amount of a good that sellers would choose to produce and sell given the price $P$, as well as other (exogenous) variables $\bar{Q_s}$, such as prices of factors of production, prices of alternative products the firm could produce, technology, productive capacity and expectations of future prices.

$$Q_s = \bar{Q_s} + bP$$

Find the value of $Q_s$ corresponding to the given price in the `price_list` and return a dictionary with the prices from `price_list` as keys and the calculated `Q_s` as values. Also assign the result of your function to dictionary `q_supply`. Your dictionary should be in the form `{2: q_s1, 9: q_s2, 10: q_s3, ...and so on}`

Assume the following values: $b$ = 5, $\bar{Q_s}$ = 250

```python
# Question 3
b = 5  # sensitivity of the supply to price
Qs_bar = 250 # exogenous variable, independent qty supplied

# your code here
def supply(price_list):
    q_s = {}
    for price in price_list:
        q_s[price] = Qs_bar + b * price

    return q_s

q_supply = supply(price_list)
q_supply
```

### Question 4: Finding Equilbrium

The equilibrium is reached when the supply and demand curves cross $Q_d = Q_s$. Determine the equilibrium price $P_e$ and the equilibrium quantity $Q_e$

$$P_e = \frac{\bar{Q_d} − \bar{Q_s}}{a+b}$$

$$Q_e = \bar{Q_d} − aP_e = \bar{Q_s} + bP_e$$

Use the above given formulae to calculate and display the equilibrium price and quantity.

```python
# Question 4

# calculate equilibrium price
P_e = (Qd_bar - Qs_bar)/(a + b)

# calculate equilibrium quantity
# it can be calulated 2 ways:
Q_e = Qd_bar - a * P_e
Q_e = Qs_bar + b * P_e

print("The equilibrium price is {} and equilibrium quantity is {}".format(P_e, Q_e))
```

### Plotting the curves!

Let's illustrate all the calculations using graph, here you will notice that the interesection point of supply and demand curves will be exactly same as coordinates calculated in Question 4

```python
### DO NOT EDIT ###
import matplotlib.pyplot as plt

# plot demand and supply curves
plt.plot(q_demand.values(), price_list, label='Demand')
plt.plot(q_supply.values(), price_list, label='Supply')
plt.scatter(Q_e, P_e, color='red', s=50, zorder=3)

# add axis labels and title
plt.ylabel('Price')
plt.xlabel('Quantity')
plt.title('Market Equilibrium')

# add legend
plt.legend()

# display plot
plt.show()
```
