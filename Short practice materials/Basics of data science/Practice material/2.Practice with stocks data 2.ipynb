## **List**
A list is a versatile data structure in Python that can store a collection of items. It is ordered and mutable, which means you can change its elements. It is very flexible as the items in a list do not need to be of the same type. This means that you can have e.g. both strings and doubles or even other lists.

A list is created by placing items inside square brackets [ ], separated by commas.

```python
# Python program to demonstrate creation of List
my_list = []
print("Intial blank List: ")
print(my_list)

# Creating a List with the use of a String
my_list = ['IntroToPython']
print("\nList with the use of String: ")
print(my_list)

# Creating a List with the use of multiple values
my_list = ["Intro", "To", "Python"]
print("\nList containing multiple values: ")
print(my_list[0])
print(my_list[2])

# Creating a Multi-Dimensional List (By Nesting a list inside a List)
my_list = [['Intro', 'To'], ['Python']]
print("\nMulti-Dimensional List: ")
print(my_list)
```

```python
# Python program to demonstrate accessing of element from list

# Creating a List with the use of multiple values
my_list = ["Python", "for", "Data", "Science"]

# accessing a element from the list using index number
print("Accessing element from the list")
print(my_list[0])
print(my_list[2])

# accessing a element using negative indexing
print("Accessing element using negative indexing")
print(my_list[-1]) # print the last element of list
print(my_list[-3]) # print the third last element of list

#accessing elements from index 1 to 3
print(my_list[1:4])
```

##Functions

A function is a reusable block of code designed to perform a specific task. It helps organize code, improves readability, and promotes reusability.

Functions follow a consistent structure:

- `function_name` represents the name of the function.
- `input1` and `input2` are parameters or inputs to the function.
- The `return` statement specifies the value the function will output, such as 1.


```python
#make the function
def function_name(input1, input2):
  #some code with using input1 and input2
  value = 1
  return value


#use the function
print("Hello world!")
print(function_name(1,2))
```

You don't have to use a return statement and you can give parameter (an input) a default value.

```python
def funtion_no_return(input1, input2):
  print(input1, input2)

funtion_no_return("Good", "Luck")

def function_default(input1, input2="world!"):
  output = input1 + input2
  return output

print(function_default("Hello ", "Kes!"))
print(function_default("Hello "))
```

*   Estimate missing values based on the values of adjacent data points.

# Functions for multiple DFs

What if we want to examine multiple stocks? In this case, we can copy and paste code multiple times, but this is not very clean, and will make it hard to debug your code if something goes wrong. Instead, we can organize the code into _functions_.

Here, we define two functions:
- `download_ticker_data`: Download ticker data and save to a file whose name is generated in the function itself (for ease of use).
- `load_ticker_data`: Load already downloaded files, and pre-process them so that they are ready to use.

```python
def download_ticker_data(ticker, period, interval):
    """ Download and save ticker data """
    df = yf.download(
            tickers = ticker,
            period = period,
            interval = interval
          )
    if isinstance(df.columns, pd.MultiIndex):
     df.columns = df.columns.get_level_values(0)
    # Create a filename by concatenating ticker, period
    # and interval with an underscore in between each.
    # Also, add a file extention at the end.
    filename = "_".join([ticker, period, interval])+".csv"
    # Save file to disk
    df.to_csv(filename)
    # print statement just to say that everything went as expected
    print("> Saved data to file:", filename, "!")

    return df

def load_ticker_data(filename):
    """ Load ticker data from CSV file already in disk """
    # This code is the exact same that we used earlier when
    # pre-processing the date column in the loaded csv file/df
    df = pd.read_csv(filename)

    df["Date"] = pd.to_datetime(df["Date"])
    df = df.set_index(df["Date"])
    df = df.drop(columns=["Date"])

    # print statement just to say that everything went as expected
    print("> Loaded file", filename, "!")

    return df
```

Thus, we can analyse the daily returns instead of the raw stocks. Assume we buy stock at the beginning of yesterday, and sell it at the beginning of today. What returns do we expect to get? For this, we can create a new column that takes the difference of the open prices on successive days:

```python
# Calculate returns
df_aapl["Returns"] = 0  # create a new column in df_aapl called Returns and set all values to 0

# iterate over all rows -- note that this is not the usual or best way to iterate
# over a dataframe, but will be clear to you now if you are familiar with for loops.
for idx in range(1, len(df_aapl.index)): # we start at the first and not zeroth row. Why?
    cur_date = df_aapl.index[idx]    # get the current row's date (remember the row's index is a date)
    prev_date = df_aapl.index[idx-1] # get the previous row's date
    cur_price = df_aapl["Open"].loc[cur_date]   # get the current row's open price
    prev_price = df_aapl["Open"].loc[prev_date] # get the previous row's open price
    df_aapl["Returns"].loc[cur_date] = cur_price-prev_price # today's return will be the difference
    # Note that we might get a warning for this last line of code. Why? How do we fix it?

# print the dataframe
df_aapl
```

Now that we have a dependable algorithm to compute returns, we can write a function to do this to any stock DataFrame we give to it:

```python
def construct_diff_col(df):
    """ Compute returns based on the open prices """
    df["Returns"] = 0
    for idx in range(1,len(df.index)):
        cur_idx = df.index[idx]
        prev_idx = df.index[idx-1]
        cur_price = df["Open"].loc[cur_idx]
        prev_price = df["Open"].loc[prev_idx]
        df["Returns"].loc[cur_idx] = cur_price-prev_price

    return df

# compute returns for both companies
df_aapl = construct_diff_col(df_aapl)
df_tsla = construct_diff_col(df_tsla)
```

We can now vizualise the returns in order to examine any patterns or gather information to create a hypothesis.

```python
# load data
df_aapl = load_ticker_data("AAPL_5y_1d.csv")
df_tsla = load_ticker_data("TSLA_5y_1d.csv")

# calculate returns
df_aapl = construct_diff_col(df_aapl)
df_tsla = construct_diff_col(df_tsla)

ax = df_tsla[["Returns"]].plot()
df_aapl[["Returns"]].plot(ax=ax)
ax.set_title("Daily returns")
ax.set_ylabel("USD")
ax.legend(["Tesla", "Apple"])
ax.grid()
plt.show()
```

### Computing volatility

Great! So now, though we saw that the Tesla stock was higher in absolute numbers, it seems like its returns jumps around a lot over time, compared to Apple's returns. This is especially true after about 2020. Thus, we can visually hypothesize that TSLA is more volatile than AAPL.

We can now go one step further: we can mathematically determine the volatility of the stocks' daily returns using the formula (which you can find in textbooks or the internet):

$$\sigma_T=\sigma\sqrt{T}$$

Where $\sigma_T$ is the volatility over time period $T$ (in this case, number of days), and $\sigma$ is the standard deviation of the returns (a measure of the range of data).

We can compute the standard deviation using the inbuilt Pandas DataFrame `std` function. We can compute the standard deviation using the inbuilt `sqrt` function from Python's own `math` library. Isn't it great not having to re-invent the wheel?

 Thus, we get:

```python
import math

std_aapl = df_aapl["Returns"].std()
std_tsla = df_tsla["Returns"].std()
n_days = len(df_aapl)

vol_aapl = std_aapl*math.sqrt(n_days)
vol_tsla = std_tsla*math.sqrt(n_days)

print("Volatility of Apple Inc:", vol_aapl)
print("Volatility of Tesla Inc:", vol_tsla)
print("Ratio (Tesla/Apple):", vol_tsla/vol_aapl)
```

Thus, we find that the find that the volatility of Tesla stock is about 3.26 times that of Apple's! So, what would _you_ invest in now?

```python
# calculate returns
def construct_diff_col(df):
    """ Compute returns based on the open prices """
    df["Returns"] = 0
    for idx in range(1,len(df.index)):
        cur_idx = df.index[idx]
        prev_idx = df.index[idx-1]
        cur_price = df["Open"].loc[cur_idx]
        prev_price = df["Open"].loc[prev_idx]
        df["Returns"].loc[cur_idx] = cur_price-prev_price

    return df

df_aapl = construct_diff_col(df_aapl)
df_tsla = construct_diff_col(df_tsla)
```

When we look at data it can be very useful to know if it comes from a certain distribution. If it comes from a certain distribution we can make assumptions about the range and likelihood of certain values. To use visual tools to determine the distribution of data you can use histograms. Histograms show the frequency of data when the x-axis is 'binned' or bucketed into several segments.


 _hist(x, bins = b, alpha = a)_. x is the data that we want to plot. b is the number of bins that we want to use, so into how many intervals we want to separate our data in and alpha is about how see through we want our histogram to be.

```python
plt.hist(df_tsla["Returns"], bins=30, alpha=0.5)
plt.hist(df_aapl["Returns"], bins=30, alpha=0.5)

```

We see that our apple return data seems to have a peak around 0 and seems to have equal weights on both sides of these peaks. It therefore seems reasonable to look if our apple returns might be normally distributed. To test this visually we can fit a a normal distribution over our data to visualize the 'expected' dataset. We can approximate the mean and standard deviation with the sample mean (μ) and sample standard deviation (σ).

$$\mu=\frac{1}{N}\sum_{i=0}^N x_i$$
$$\sigma=\sqrt{\frac{\sum_{i=0}^N (x_i-\mu)^2}{N}}$$

If we know that we expect a normal distribution, we can fit it using the computed mean and standard deviation by putting it into the function:

$$\mathcal{N}(x|\mu,\,\sigma^{2})=\frac{1}{\sigma\sqrt{2\pi}}\exp\left(-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2\right)$$

```python
import numpy as np
import scipy.stats as stats

#new_df = df_aapl.loc[(5>df_aapl["Returns"]) & (df_aapl["Returns"]>-5)] #only keep returns that are between -5 and 5

new_df = df_aapl

m = new_df["Returns"].mean() #mean of returns from new_df
s = new_df["Returns"].std() #standard deviation of returns from new_df

x = np.linspace(new_df["Returns"].min(), new_df["Returns"].max(), 100) #creates a 100 numbers evenly spaced between the min and max
y = stats.norm.pdf(x, m, s)
plt.hist(new_df["Returns"], bins=30, density=True)
plt.plot(x, y)
plt.show()
```

```python
#plot the normal distribution fitted to the apple's and tesla's returns

m_a = df_aapl["Returns"].mean()
s_a = df_aapl["Returns"].std()

m_t = df_tsla["Returns"].mean()
s_t = df_tsla["Returns"].std()

x = np.linspace(new_df["Returns"].min(), new_df["Returns"].max(), 100)
y_a = stats.norm.pdf(x, m_a, s_a)
y_t = stats.norm.pdf(x, m_t, s_t)

plt.plot(x,y_a, label="Apple dist")
plt.plot(x,y_t, label="Tesla dist")

plt.legend()
plt.show()
```

## T-Tests

A t-test is used to determine whether the difference of means of two distributions is zero. H0: Null hypothesis: "The distribution means of the two datasets are the same."

If the p-value is less than some threshold, usually 0.05, we can reject the null hypothesis. Else, we cannot reject it (note that this is _not_ the same as 'accepting' it).

Note that, in general, we have these four assumptions:
1. The data is continuous.
2. The sample data is randomly sampled from some underlying population.
3. Homogeneity of variance (the variance in both datasets are similar).
4. The sample distribution is approximately normal.

There are variants of the t-tests if these assumptions do not hold.

(Perfect time to Google! Example: https://www.investopedia.com/terms/t/t-test.asp)

If we are deciding between two stocks then it might also be good to see if the returns are significantly different or not. We can do this with the following code. Since the p value is 0.80 we can not reject the null hypotheses of similarity.

```python
stats.ttest_ind(df_tsla["Returns"], df_aapl["Returns"]) # Type I , II
```
